# Road Accident Data Analysis

## Objective
The objective of this project is to analyze road accident data to identify patterns, trends, and factors affecting accidents, injuries, and deaths.

## Datset:
Source:Data.gov.in

## Importing Required Libraries

 We import necessary libraries for data analysis and visualization.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading the Dataset
The dataset is loaded using pandas for further analysis.

```python
df=pd.read_csv("State_Wise_RoadAccident.csv")
```

## Understanding the Data
We examine the structure, data types, and summary of the dataset.
Checking Missing Values

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
# replacing "- and " " " into "_"
df.columns = df.columns.str.replace(' ', '_')
df.columns= df.columns.str.replace('-','_')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
# handling datatypes
df['Curved_Road___Persons__Killed'] = df['Curved_Road___Persons__Killed'].astype(int)

df["Bridge___Persons__Killed"] = df["Bridge___Persons__Killed"].astype(int)

df["Ongoing_Road_Works/Under_Construction___Persons__Killed"] = df["Ongoing_Road_Works/Under_Construction___Persons__Killed"].astype(int)
df["Year"]=df["Year"].astype(int)
```

```python
df.info()
```

```python
df.describe()
```

## Feature Engineering
New columns such as Total Accidents, Total Deaths, and Total Injured are created by combining relevant columns.

```python
accident_cols = [col for col in df.columns if "Accident" in col]
df["Total Accidents"] = df[accident_cols].sum(axis=1)

injured_cols = [col for col in df.columns if "Total_Injured" in col]
df["ToTal Injured"] = df[injured_cols].sum(axis=1)

Killed_cols = [col for col in df.columns if "Killed" in col]
df["Total Deaths"] = df[Killed_cols].sum(axis=1)
```

## State-wise Analysis
We identify top states based on accidents, deaths, and injuries.

```python
state_data = df.groupby('State/UT')['Total Accidents'].sum().sort_values(ascending=False)

state_data.head(10).plot(kind='bar')
plt.title("Top 10 States by Accidents")
plt.show()
```

### Insight:
* Tamil Nadu has the highest number of accidents, followed by Madhya Pradesh and Uttar Pradesh. 
* Karnataka and Kerala also report high accident counts. 
* This shows that southern and central states contribute significantly to total accidents, likely due to higher traffic density and urbanization.

```python
state_data = df.groupby('State/UT')['ToTal Injured'].sum().sort_values(ascending=False)

state_data.head(10).plot(kind='bar')
plt.title("Top 10 States by Injured")
plt.show()
```

### Insight:
* Tamil Nadu again records the highest number of injured persons, followed by Madhya Pradesh and Karnataka.
* Kerala and Uttar Pradesh also show high injury counts.
*  This confirms a strong relationship between accident frequency and number of injuries.

```python
state_data = df.groupby('State/UT')['Total Deaths'].sum().sort_values(ascending=False)

state_data.head(10).plot(kind='bar')
plt.title("Top 10 States by Deaths")
plt.show()
```

### Insight:
* Uttar Pradesh has the highest number of deaths, significantly higher than other states, followed by Maharashtra and Madhya Pradesh.
* Tamil Nadu and Karnataka also contribute notably.
*  This indicates that Uttar Pradesh experiences more severe or fatal accidents.

## Conclusion:
States like Tamil Nadu, Madhya Pradesh, and Uttar Pradesh consistently appear across accidents, injuries, and deaths, highlighting them as high-risk regions.
While Tamil Nadu leads in accidents and injuries, Uttar Pradesh shows the highest fatalities, indicating higher accident severity in that state.

## Year-wise Analysis
We analyze trends in accidents, injuries, and deaths over time.

```python
year_data = df.groupby('Year')['Total Accidents'].sum().reset_index()

sns.barplot(x='Year', y='Total Accidents', data=year_data)
plt.title("Year wise Total Accidents")
plt.show()
```

### Insight:
* The total number of accidents is highest in 2019, drops significantly in 2020, and increases again in 2021.
* The decline in 2020 may be due to reduced travel and mobility, while the rise in 2021 indicates a return toward normal traffic conditions.

```python
year_data = df.groupby('Year')['ToTal Injured'].sum().reset_index()

sns.barplot(x='Year', y='ToTal Injured', data=year_data)
plt.title("Year wise  Total Injured")
plt.show()
```

### insights
* The total number of injured is highest in 2019, drops significantly in 2020, and increases again in 2021.
*  2021 indicates a Partially recover(still below 2019)

```python
year_data = df.groupby('Year')['Total Deaths'].sum().reset_index()

sns.barplot(x='Year', y='Total Deaths', data=year_data)
plt.title("Year wise  Total Deaths")
plt.show()
```

### insights
* Lower in 2020 than 2019 The total number of deaths lower in 2020 than 2019
* it incrase in 2021 more than 2019

## Relationship Analysis
We analyze the relationship between accidents and deaths.

```python
# Compare Accidents vs Deaths
sns.scatterplot(x='Total Accidents', y='Total Deaths', data=df)
plt.title("Accidents vs Deaths")
plt.show()
```

## Death Rate Analysis
Death rate is calculated to understand accident severity.

```python
# Death rate over time
year_data = df.groupby('Year')[['Total Deaths','Total Accidents']].sum()
year_data['Death Rate'] = year_data['Total Deaths'] / year_data['Total Accidents']
year_data['Death Rate'].plot()

plt.title("Death Rate Over Time")
plt.xlabel("Year")
plt.ylabel("Death Rate")
plt.xticks(year_data.index)

plt.show()
```

### insights
* Increases from 2019 to 2020
* Continues rising in 2021

```python
State_data = df.groupby('State/UT')[['Total Deaths','Total Accidents']].sum()
State_data['Death Rate'] = State_data['Total Deaths'] / State_data['Total Accidents']
top10 = State_data.sort_values('Death Rate', ascending=False).head(10)
top10['Death Rate'].plot(kind="barh")

plt.title("Death Rate by state")
plt.xlabel("State")
plt.ylabel("Death Rate")
plt.show()
```

## Key Insights

- Accidents are concentrated in a few states.
- There is a strong relationship between accidents and deaths.
- Year-wise trends show variation due to external factors.
- Death rate helps identify severity of accidents.

This analysis highlights key trends and patterns in road accidents.
